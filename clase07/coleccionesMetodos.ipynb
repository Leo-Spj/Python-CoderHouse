```python
# el peor metodo de ordenamiento:

def ordenamientoBurbuja(lista):
    for i in range(len(lista)):
        for j in range(len(lista)-1):
            if lista[j] > lista[j+1]:
                lista[j], lista[j+1] = lista[j+1], lista[j]

# crear una lista de numeros aleatorios con 80000 elementos
import random
lista = [random.randint(0,100) for i in range(8000)]

ordenamientoBurbuja(lista)
print(lista)
```

```python
# Metodo isdisjoint

conjunto = {1, 2, 3, 4, 5}
conjunto2 = {6, 7, 8, 9}
conjunto3 = {5, 6, 7, 8}
print(conjunto.isdisjoint(conjunto2))
print(conjunto.isdisjoint(conjunto3))
```

```python
#Metodo issubset

conjunto = {1, 2, 3, 4, 5}
conjunto2 = {1, 2, 3}
print(conjunto2.issubset(conjunto)) # conjunto2 es subconjunto de conjunto - true
print(conjunto.issubset(conjunto2)) # conjunto es subconjunto de conjunto2 - false
```

```python
# Metodo issuperset

conjunto = {1, 2, 3, 4, 5}
conjunto2 = {1, 2, 3}
print(conjunto2.issuperset(conjunto)) # conjunto2 es superconjunto de conjunto - false
print(conjunto.issuperset(conjunto2)) # conjunto es superconjunto de conjunto2 - true
```

```python
# Metodo union

conjunto = {1, 2, 3, 4, 5}
conjunto2 = {1, 2, 3}
print(conjunto.union(conjunto2)) # union de conjuntos
```

```python
# Metodo difference

conjunto = {1, 2, 3, 4, 5}
conjunto2 = {1, 2, 3}
print(conjunto.difference(conjunto2)) # diferencia de conjuntos
```

```python
# Metodo intersection

conjunto = {1, 2, 3, 4, 5}
conjunto2 = {1, 2, 3}
print(conjunto.intersection(conjunto2)) # interseccion de conjuntos
```

```python
# Metodo intersection_update

conjunto = {1, 2, 3, 4, 5}
conjunto2 = {1, 2, 3}
conjunto.intersection_update(conjunto2) # interseccion de conjuntos, hace el update en el conjunto, 
# en otras palabras, el conjunto queda con la interseccion de los dos conjuntos y el otro conjunto queda igual
print(conjunto) # interseccion de conjuntos
```
